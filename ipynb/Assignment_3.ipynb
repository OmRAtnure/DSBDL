```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/drive/MyDrive/Datasets/data.csv')
df.head()
```

```python
grouped_stats = df.groupby("gender")["income"].agg(
    mean="mean",
    median="median",
    minimum="min",
    maximum="max",
    std_dev="std"
)

grouped_stats
```

```python
income_list_by_gender = df.groupby("gender")["income"].apply(list)

income_list_by_gender
```

```python
# 2.Write a Python program to display some basic statistical details like percentile, mean, standard
#deviation etc. of the species of ‘Iris-setosa’, ‘Iris-versicolor’ and ‘Iris- versicolor’ of iris.csv
#dataset.
```

```python
iris = pd.read_csv("/content/drive/MyDrive/Datasets/Iris.csv")
iris.head()
```

```python
print(iris.columns)
```

```python
iris_stats = iris.groupby("Species").describe()
iris_stats
```

```python
for species in iris["Species"].unique():
    print(f"\nStatistics for {species}")

    subset = iris[iris["Species"] == species]

    print("Mean:\n", subset.mean(numeric_only=True))
    print("\nStandard Deviation:\n", subset.std(numeric_only=True))
    print("\nPercentiles:\n", subset.quantile([0.25, 0.50, 0.75], numeric_only=True))
```
